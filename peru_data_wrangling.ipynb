```python
import pandas as pd
```

```python
peru_field1 = pd.read_csv('cleaned_data/BDF_2019_1_expanded.tsv', sep='\t')
peru_field1.head()
```

```python
peru_field1 = peru_field1.rename(columns=lambda col: col.rstrip())
```

```python
peru_field1['file_name'] = peru_field1['file_name'] + '.JPG'
```

```python
peru_field3 = pd.read_csv('cleaned_data/BDF_2019_3_expanded.tsv', sep='\t')
peru_field3.head()
```

```python
peru_field = pd.concat([peru_field1, peru_field3])
peru_field.head()
```

```python
peru_field.to_csv('cleaned_data/peru_field_combined.tsv',
                  sep='\t', index=False)
```

```python
peru_field_cols = {'file_name':'file_name',
                   'Genero':'genus',
                   'Especie':'species',
                   'Código/ ID del individuo*':'specimen_id'}
peru_field = peru_field.rename(columns=peru_field_cols)
peru_field['data_source'] = 'field'
peru_field['country'] = 'Peru'

peru_field = peru_field[['file_name','genus','species','data_source',
                         'country','specimen_id']]
peru_field.head()
```

```python
museum_df = pd.read_csv('original_data/Museum_Sheets/Museum_Peru_2020.csv')
museum_df.head()
```

```python
museum_df.info()
```

```python
museum_df['catalogNumber'] = museum_df['catalogNumber'].astype('str')
```

```python
museum_cols = {'photoID':'file_name',
                'specificEpithet':'species'}
museum_df = museum_df.rename(columns=museum_cols)
museum_df['specimen_id'] = 'USNM:Fish:' + museum_df['catalogNumber']
museum_df['data_source'] = 'museum'
museum_df = museum_df[['file_name','genus','species','data_source',
                         'country','specimen_id']]
museum_df.head()
```

```python
combined = pd.concat([peru_field, museum_df])
combined.head()
```

```python
combined.to_csv('fish_image_metadata_combined.csv', index=False)
```
