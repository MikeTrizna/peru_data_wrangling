```python
import pandas as pd
```

```python
peru_field1 = pd.read_csv('original_data/Field_Sheets/BDF_2019_1.csv')
peru_field1.head()
```

```python
peru_field1.info()
```

```python
expanded_rows = []
for index, row in peru_field1.iterrows():
    row_dict = row.to_dict()
    codigo = row_dict['Codigo foto']
    if ' DSC' in codigo:
        codigo_split = codigo.split(' ')
        row_dict['file_name'] = codigo_split[0]
        expanded_rows.append(row_dict)
        row_dict['file_name'] = codigo_split[1]
        expanded_rows.append(row_dict)
    elif '-DSC' in codigo:
        codigo_split = codigo.split('-DSC')
        row_dict['file_name'] = codigo_split[0]
        expanded_rows.append(row_dict)
        row_dict['file_name'] = f'DSC{codigo_split[1]}'
        expanded_rows.append(row_dict)
    else:
        row_dict['file_name'] = codigo
        expanded_rows.append(row_dict)
expanded_df = pd.DataFrame(expanded_rows)
```

```python
expanded_df.info()
```

```python
expanded_df.to_csv('cleaned_data/BDF_2019_1_expanded.tsv',
                   sep='\t', index=False)
```
